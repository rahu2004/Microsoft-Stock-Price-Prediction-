```python
# Prepare the training set samples
msft_close = microsoft.filter(['close'])
dataset = msft_close.values
training = int(np.ceil(len(dataset) * 0.95))
# Scale the data
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
dataset = ss.fit_transform(dataset)

train_data = dataset[0:int(training), :]

x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
X_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

# **Build the Model**

```python
model = keras.models.Sequential()
model.add(keras.layers.LSTM(units=64,
							return_sequences=True,
							input_shape
							=(X_train.shape[1], 1)))
model.add(keras.layers.LSTM(units=64))
model.add(keras.layers.Dense(128))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(1))

print(model.summary())
```

# **Compile and Fit**

```python
from keras.metrics import RootMeanSquaredError
model.compile(optimizer='adam',
			loss='mae',
			metrics=RootMeanSquaredError())

history = model.fit(X_train, y_train,
					epochs=20)
```

# **Model Evaluation**

```python

testing_data = dataset[training - 60:, :]
testing_data = ss.transform(testing_data)  # Use transform, not direct indexing

x_test = []
y_test = dataset[training:, :]  # Assuming you have ground truth values for evaluation

# Create the x_test data using the same sliding window approach as in the training set
for i in range(60, len(testing_data)):
    x_test.append(testing_data[i - 60:i, 0])

x_test = np.array(x_test)
X_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

# Make predictions using the trained model
pred = model.predict(X_test)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)

# You can also inverse transform the predictions if you want to get them back to the original scale
pred_original_scale = ss.inverse_transform(pred)
```
